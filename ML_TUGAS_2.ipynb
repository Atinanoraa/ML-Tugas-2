#9. Ubah angka pada kolom Sales sesuai dengan ketentuan berikut :
*   Pelanggan yang tinggal di “South” mendapatkan diskon 10%
*  Pelanggan yang tinggal di “West” mendapatkan diskon 15%
*   Pelanggan yang tinggal di “Central” mendapatkan diskon 13%
*   Pelanggan yang tinggal di “East” mendapatkan diskon 20!


```python
def calculate_discount(region, sales):
    if region == 'South':
        return 0.9 * sales  # Diskon 10%
    elif region == 'West':
        return 0.85 * sales  # Diskon 15%
    elif region == 'Central':
        return 0.87 * sales  # Diskon 13%
    elif region == 'East':
        return 0.8 * sales  # Diskon 20%
    else:
        return sales  # Tidak ada diskon

data['Discounted_Sales'] = data.apply(lambda row: calculate_discount(row['Region'], row['Sales']), axis=1)
```
